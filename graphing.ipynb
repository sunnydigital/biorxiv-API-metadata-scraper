# Notebook to Graph Outputs

---

This current notebook has to do with graphing the outputs of the scraped information from bioRxiv

```python
import matplotlib.pyplot as plt
import json
```

```python
with open('global_dicts/return_dict.json', 'r') as f:
    return_dict = json.load(f)
```

```python
license_data = return_dict['license']
license_data
```

```python
plt.pie(license_data.values(), labels=license_data.keys(), autopct='%1.1f%%')
plt.title('bioRxiv license distribution')
```
